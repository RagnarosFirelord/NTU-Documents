```python
iris=datasets.load_iris()
iris_data= pd.DataFrame(iris.data,columns=iris.feature_names)
iris_data['species']=iris.target_names[iris.target]
# iris_data.head(3).append(iris_data.tail(3)) 
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
pd.set_option('max_colwidth',50)
iris_data
```

```python
def scalarization(data, k):
    _range = k
    return (data - np.min(data)) / _range
```

```python
def normalization(data):
    return (data-np.min(data))/(np.max(data)-np.min(data))
```

```python
trainX=iris_data.loc[:,["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]]
trainX
```

```python
trainY=iris_data['species']
trainY
```

```python
trainX['petal width (cm)']=normalization(iris_data['petal width (cm)'])
trainX['petal length (cm)']=normalization(iris_data['petal length (cm)'])
trainX['sepal width (cm)']=normalization(iris_data['sepal width (cm)'])
trainX['sepal length (cm)']=normalization(iris_data['sepal length (cm)'])
clf = neighbors.KNeighborsClassifier(7, weights='distance')
clf.fit(trainX, trainY)
scores = cross_val_score(clf, trainX, trainY, cv=6, scoring='accuracy')
print(scores.mean())
```
